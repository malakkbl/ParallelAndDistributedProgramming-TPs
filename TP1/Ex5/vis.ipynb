```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV (same folder)
df = pd.read_csv("hpl_results.csv")

# Keep only successful runs
df = df[df["status"] == "PASSED"]
```

```python
plt.figure()

for N in sorted(df["N"].unique()):
    data = df[df["N"] == N]
    plt.plot(data["NB"], data["gflops"], marker='o', label=f"N={N}")

plt.xlabel("Block size (NB)")
plt.ylabel("Performance (GFLOP/s)")
plt.title("HPL Performance vs Block Size")
plt.legend()
plt.grid(True)

plt.savefig("hpl_gflops_vs_nb.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
plt.figure()

for N in sorted(df["N"].unique()):
    data = df[df["N"] == N]
    plt.plot(data["NB"], data["time_s"], marker='o', label=f"N={N}")

plt.xlabel("Block size (NB)")
plt.ylabel("Execution time (s)")
plt.title("HPL Execution Time vs Block Size")
plt.legend()
plt.grid(True)

plt.savefig("hpl_time_vs_nb.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
PEAK_GFLOPS = 17.6  # <-- change if needed

df["efficiency"] = df["gflops"] / PEAK_GFLOPS

plt.figure()

for N in sorted(df["N"].unique()):
    data = df[df["N"] == N]
    plt.plot(data["NB"], data["efficiency"], marker='o', label=f"N={N}")

plt.xlabel("Block size (NB)")
plt.ylabel("Efficiency")
plt.title("HPL Efficiency vs Block Size")
plt.legend()
plt.grid(True)

plt.savefig("hpl_efficiency_vs_nb.png", dpi=300, bbox_inches="tight")
plt.show()
```
